Import necessary libraries

```python
import boto3
import json
import base64
import time # Import the time module for timing
```

The end-point was crated with US-West-2 region in AWS which corresponds to Oregon

```python

# Initialize SageMaker runtime client with region specified
# Initialize SageMaker runtime client with explicit credentials
sagemaker_runtime = boto3.client(
    'sagemaker-runtime',
    #removed access keys
    region_name='us-west-2',
    config=boto3.session.Config(read_timeout=1000)
    )
```

check image size and resize

```python
from PIL import Image
import matplotlib.pyplot as plt



# Load the image
image = Image.open(image_path)

# Check image size
original_size = image.size
print(f"Original size: {original_size}")

# Plot the image
plt.imshow(image)
plt.axis('off')
plt.title('Original Image')
plt.show()
```

resize from 640, 480 to smaller 200

```python
# Set maximum size (width, height)
max_size = (400, 300)
# Resize if the image is too large
if image.size[0] > max_size[0] or image.size[1] > max_size[1]:
  image.thumbnail(max_size, Image.LANCZOS)
  print(f"Resized image to: {image.size}")

# Plot the image
plt.imshow(image)
plt.axis('off')
plt.title('Resized Image')
plt.show()
```

save image

```python
# Save the resized image to local Colab drive
output_image_path = '/content/TranSalNet/example/reduced_COCO_val2014_000000005107.jpg'
image.save(output_image_path)
print(f"Resized image saved to {output_image_path}")

# Plot the image
plt.imshow(image)
plt.axis('off')
plt.title('Resized Image')
plt.show()
```

Create the paylod

```python
with open(output_image_path, 'rb') as image_file:
    encoded_image = base64.b64encode(image_file.read()).decode('utf-8')
```

```python
# Create payload
payload = json.dumps({'image': encoded_image})

```

whats in payload?

```python
print(payload)
```

```python

end_point ='pytorch-serverless-endpoint'
```

Try and add a timer this time

```python
from datetime import datetime, timezone
start_time = time.time()
try:
    # Invoke endpoint
    response = sagemaker_runtime.invoke_endpoint(
        EndpointName=end_point,
        ContentType='application/json',
        Body=payload
    )

    # Calculate response time
    end_time = time.time()  # End timer
    response_time = end_time - start_time
    print(f"Response time: {response_time} seconds")

    # Read response
    result = json.loads(response['Body'].read())
    #print(result)

except Exception as e:
    # Calculate response time in case of an exception
    end_time = time.time()
    response_time = end_time - start_time
    print(f"Invocation failed, response time: {response_time} seconds")
    print(f"Error: {e}")
    print(datetime.fromtimestamp(end_time, tz=timezone.utc))
```
